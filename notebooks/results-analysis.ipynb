```python
import os
import matplotlib.pyplot as plt
import pandas as pd



df=None
for a,b,allfiles in os.walk('../results/'):
    for c in allfiles:
        if not (c[:6] == '6years'): continue
        if 'Only' in c: 
            print('Only')
            continue
        cdf = pd.read_csv('../results/'+c, header=1)

        if df is None: df=cdf
        else:
            df = pd.concat([df,cdf])

keys = df.keys().to_list()
print(keys)
figcols = 1

fig,ax = plt.subplots(11,figcols,figsize=(10,20))
fig.tight_layout(pad=2)

for i in range(11):
    key = keys[i+1]
    curr_ax = ax[i//figcols]#[i%figcols]
    curr_ax.scatter(df[key],[0]*len(df[key]))
    curr_ax.set_title(key)
    curr_ax.set_xlim(-0.4,0.4)
    if key=='bias':
        curr_ax.set_xlim(0.0,1.2)
        curr_ax.set_facecolor('#E0E0E0')
        break
        #the rest of the variables are used too rarely
    

fig.show()
fig.savefig('../Results.png')
```
